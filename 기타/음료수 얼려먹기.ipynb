```python
서로 인접한 노드형태로 표현하여 dfs,bfs로 푼다
```

```python
n,m=4,5

li=[0,0,0,0]
li[0]=[0,0,1,1,0]
li[1]=[0,0,0,1,1]
li[2]=[1,1,1,1,1]
li[3]=[0,0,0,0,0]
```

```python
def dfs(x,y):
    if x < 0 or y < 0 or y >= m or x >= n:
        return False
    
    if li[x][y]==0:
        li[x][y]=1
        
        dfs(x,y-1)
        dfs(x,y+1)
        dfs(x-1,y)
        dfs(x+1,y)
        return True
    return False
```

```python
count=0
for i in range(n):
    for j in range(m):
        if dfs(i,j)==True:
            count += 1
print(count)
```
