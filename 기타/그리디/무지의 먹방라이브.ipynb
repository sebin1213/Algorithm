```python
def solution(food_times, k):
    food_len = len(food_times)
    zero,count,i = 0,0,0
    if food_len > k:
        return (k+1)
    while (count-1) != k:        
        if zero == food_len:
            return -1
        if i >= food_len:
            i-=food_len
        if food_times[i] > 0:
            food_times[i] -= 1
            count += 1
            zero = 0
        else:
            zero +=1
        i+=1
    return i
```
