```python
from itertools import permutations

n = int(input())
li = list(map(int,input().split()))

max_val=0
value=0
for i in list(permutations(li)):
    value=0
    for j in range(1,n):
        value += abs(i[j]-i[j-1])
    max_val=max(max_val,value)
print(max_val)
```
